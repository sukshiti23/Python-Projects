Calculating the track and it's artist with most and least streams in 2023

```python
#MOST STREAMS
max_streams=data['streams'].max
artist_max_streams=data.loc[data['streams'].idxmax(),'artist(s)_name']
print("Artist Name: ",artist_max_streams)
track_max_streams=data.loc[data['streams'].idxmax(),'track_name']
print("Track Name: ",track_max_streams)
```

```python
#MIN STREAMS
min_streams=data['streams'].min
artist_min_streams=data.loc[data['streams'].idxmin(),'artist(s)_name']
print("Artist Name: ",artist_max_streams)
track_min_streams=data.loc[data['streams'].idxmin(),'track_name']
print("Track Name: ",track_min_streams)
```

Number of Tracks Per Artist

```python
count_tracks_artist=data.groupby('artist(s)_name')['track_name'].count()
print(count_tracks_artist)
print("Maximum Number of Tracks by an Artist: ",count_tracks_artist.max())
```

Artist having maximum number of tracks

```python
artist_max_tracks=count_tracks_artist.idxmax()
print("Artist with highest number of tracks: ",artist_max_tracks)
```

Using a pie chart we can determine the years of the songs which were streamed in 2023

```python
'''We grouped the release year of the songs in one variable and showed the release year of the songs which were streamed in 2023'''
yrs_grouped=data.groupby('released_year')['released_year'].value_counts().tail(10)
plt.figure(figsize=(10,8))
plt.pie(yrs_grouped,labels=yrs_grouped.index,autopct='%1.1f%%',colors=['red','blue','green','yellow','pink','orange','hotpink','lightblue','brown','cyan'])
plt.title("Distribution of Streams by Song Release Year (Spotify 2023)")
plt.show()
```

Calculating mean and median of spotify streams in 2023

Interpretation: Some of the most streamed songs in 2023 were from 2022, followed by 2023 and 2021. Most people were listening to the recent songs especially those from 2022 and 2023 , while fewers songs which were streamed were from previosu years. This indicates new music trends in the industry.

```python
data['streams']=pd.to_numeric(data['streams'],errors='coerce')
avg_streams=data['streams'].mean()
print("Average Streams = ",avg_streams)
```

```python
med_streams=data['streams'].median()
print("Median of Streams = ",med_streams)
```

Top 5 Songs and Artists

```python
top_5_songs = data.nlargest(5, 'streams')[['artist(s)_name', 'track_name']]
print(top_5_songs)
```

Top 10 tracks in the apple playlist and visualize the distribution

```python
top_10_tracks=data.nlargest(10,'in_apple_charts')[['track_name','in_apple_charts']].sort_values(by='in_apple_charts', ascending=False)
print(top_10_tracks)

plt.figure(figsize=(8,6))
plt.bar(top_10_tracks['track_name'],top_10_tracks['in_apple_charts'],color='blue')
plt.xticks(rotation=45, ha='right') 
plt.xlabel("Track Name")
plt.ylabel("Performance in Apple Charts")
plt.title("Top 10 Tracks in Apple Charts")
plt.show()
```

Top 10 artists in apple playlists and visualize

```python
top_10_artist=data.nlargest(10,'in_apple_playlists')[['artist(s)_name','in_apple_playlists']].sort_values(by='in_apple_playlists', 
                                                                                                          ascending=False)
print(top_10_artist)

plt.figure(figsize=(8,6))
plt.bar(top_10_artist['artist(s)_name'],top_10_artist['in_apple_playlists'],color='orange')
plt.xticks(rotation=45, ha='right') 
plt.xlabel("Artist Name")
plt.ylabel("Performance in Apple Playlists")
plt.title("Top 10 Artists in Apple Playlists")
plt.show()
```

Using a pair-plot to determine how danceability and BPM impact the streams.


```python
import seaborn as sns
```

```python
sns.pairplot(data[['streams', 'bpm', 'danceability_%','released_year']], diag_kind="kde",hue='released_year')
plt.show()
```

Interpretation: Most songs in the dataset are from recent years and while songs have different BPM and danceability, only a few get very high stream counts.

Monthly Streams and Visualize

```python
monthly_streams=data.groupby('released_month')['streams'].sum().sort_index(ascending=True)
print(monthly_streams)
```

```python
plt.figure(figsize=(8,6))
plt.bar(monthly_streams.index,monthly_streams.values,color='green')
plt.title("Month-Wise Streams")
plt.xlabel("Month")
plt.ylabel("Streams")
'''using xticks for displaying the months'''
plt.xticks(range(1,13), ["Jan", "Feb", "Mar", "Apr", "May", "Jun","Jul", "Aug", "Sep", "Oct", "Nov", "Dec"])
plt.show()
```

Interpretation: January saw the highest number of streams on Spotify followed by May and November. February saw a drop in the 
streams which suggests fewer high-performing releases.
